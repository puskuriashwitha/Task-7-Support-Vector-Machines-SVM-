```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd
```

```python
 # Load Dataset
df = pd.read_csv("/content/breast-cancer.csv")
print(df.head())
```

```python
# 1. Load and prepare a dataset (we'll use a 2D subset of the Iris dataset for visualization)
iris = datasets.load_iris()
X = iris.data[:, :2]  # Use only the first two features for 2D visualization
y = iris.target
```

```python
 #Convert to binary classification (e.g., class 0 vs. class 1)
X = X[y != 2]
y = y[y != 2]
```

```python
# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 2. Train SVM with linear and RBF kernel
svm_linear = SVC(kernel='linear', C=1)
svm_rbf = SVC(kernel='rbf', C=1, gamma=0.5)

svm_linear.fit(X_train, y_train)
svm_rbf.fit(X_train, y_train)
```

```python
# 3. Visualize decision boundary
def plot_decision_boundary(model, X, y, title):
    h = 0.02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
```

```python
plot_decision_boundary(svm_linear, X_train, y_train, "SVM with Linear Kernel")
plot_decision_boundary(svm_rbf, X_train, y_train, "SVM with RBF Kernel")
```

```python
# 4. Tune hyperparameters using GridSearchCV
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.5, 0.1, 0.01],
    'kernel': ['rbf']
}
```

```python
grid = GridSearchCV(SVC(), param_grid, refit=True, cv=5)
grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
```

```python
# 5. Evaluate performance
y_pred = grid.predict(X_test)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Visualize best model
plot_decision_boundary(grid.best_estimator_, X_train, y_train, "Best RBF SVM Model")
```
